# Imports

```python
import os
import zipfile
from pdf2image import convert_from_path, convert_from_bytes
import pytesseract
from PIL import Image
import re
import cv2
import numpy
import aspose.words as aw
```

# All funtions

```python
def pdfreader(path):
    images = convert_from_path(path)
    return images
```

```python
def textextracter(im):
    extractedInformation = pytesseract.image_to_string(im)
    text = str(extractedInformation)
    txt = f'{text}'
    final = txt.lower()
    return final
```

```python
def checkaadhar(final):
    n = 0
    regex = ("^[2-9]{1}[0-9]{3}\\" +"s[0-9]{4}\\s[0-9]{4}$")
    p = re.compile(regex)
    li = final.split('\n')

    if 'aadhaar' in final:
        print("\nAadhar card is present")
        print("\n",final)
        return True
    elif 'uidai' in final:
        print("\n",final)
        print("\nAadhar card is present")
        return True
    else:
        return False
```

```python
def checkpan(final):
    if "income" in final or "tax" in final or "department" in final or 'permanent account number' in final and 'salary' not in final and 'certificate' not in final:
        print("\nPan card is present")
        print("\n",final)
        return True
    else:
        return False
```

```python
def checkdriver(final):
    if 'vehicles' in final:
        print("\nDriver licsense is present")
        print("\n",final)
        return True
    else:
        return False
```

```python
def checkvoter(final):
    if 'elector' in final or 'election' in final:
        print("\nVoter card is present")
        print("\n",final)
        return True
    else:
        return False
```

```python
def checkbank(final):
    if 'statement' in final:
        print("\nBank Statement is present")
        print("\n",final)
        return True
    else:
        return False
```

```python
def checkbill(final):
    if 'water' in final:
        print("\nwater bill is present")
        print("\n",final)
        return True
    else:
        return False
```

```python
def checkpassport(final):
    if 'republic' in final:
        print("\npassport is present")
        print("\n",final)
        return True
    else:
        return False
```

```python
def checksalary(final):
    if 'salary' in final:
        print("\nsalary slip is present")
        print("\n",final)
        return True
    else:
        return False
```

```python
def checkform16(final):
    if 'form no. 16' in final:
        print("\nFORM NO 16 is present")
        print("\n",final)
        return True
    else:
        return False
```

```python
def checkcheque(final):
    if 'rupees' in final:
        print("\nCheque is present")
        print("\n",final)
        return True
    else:
        return False
```

# Main code

```python
#main
def classify(foldername):

    n = 10**6

    ogpath = '/content'
    path = os.path.join(ogpath,foldername)

    pdf = 0
    pic = 0
    docu = 0
    flag = 0

    aadhar = 0
    pan = 0
    driver = 0
    voter = 0
    bank = 0
    bill = 0
    cheque = 0
    passport = 0
    salary = 0
    form16 = 0

    aadharpath = ''
    panpath = ''
    driverpath = ''
    voterpath = ''
    bankpath = ''
    billpath = ''
    chequepath = ''
    passportpath = ''
    salarypath = ''
    newpath = ''
    form16path = ''

    for file in os.listdir(path):
        if '.pdf' in file:
            pdf += 1
            flag += 1
            pdfpath = os.path.join(path,file)
        elif '.doc' in file:
            docu += 1
            flag += 1
            docpath = os.path.join(path,file)
        elif '.png' in file:
            pic += 1
            flag += 1
            picpath = os.path.join(path,file)
        elif '.jpeg' in file:
            flag += 1
            pic += 1
            picpath = os.path.join(path,file)
        elif '.jpg' in file:
            pic += 1
            flag += 1
            picpath = os.path.join(path,file)
        else:
            flag = 0
            print('Add a file in valid format in userinput folder')

        if (flag > 0):

            for i in range(n):
                if f'classifieddocs{i}' not in os.listdir(ogpath):
            
                    newdir = f'classifieddocs{i}'
                    newpath = os.path.join(ogpath,newdir)
                    os.mkdir(newpath)
                    print('Check output in folder ',newpath)
                    if (pdf > 0):
                        images = pdfreader(pdfpath)
                        for im in images:
                            final = textextracter(im)
                            if (final.isspace() == False):
                                form16, form16path, aadhar,aadharpath,pan,panpath,driver,driverpath,voter,voterpath,bank,bankpath,bill,billpath,cheque,chequepath,passport,passportpath,salary,salarypath = classifier(final,newpath,im,form16, form16path, aadhar,aadharpath,pan,panpath,driver,driverpath,voter,voterpath,bank,bankpath,bill,billpath,cheque,chequepath,passport,passportpath,salary,salarypath)
                                
                    if(docu > 0):
                        doc = aw.Document(docpath)
                        doc.save("/content/converted.pdf")
                        images1 = pdfreader('/content/converted.pdf')
                        for im in images1:
                            final = textextracter(im)
                            if (final.isspace() == False):
                                form16, form16path, aadhar,aadharpath,pan,panpath,driver,driverpath,voter,voterpath,bank,bankpath,bill,billpath,cheque,chequepath,passport,passportpath,salary,salarypath = classifier(final,newpath,im,form16, form16path, aadhar,aadharpath,pan,panpath,driver,driverpath,voter,voterpath,bank,bankpath,bill,billpath,cheque,chequepath,passport,passportpath,salary,salarypath)
                    break
        
```
